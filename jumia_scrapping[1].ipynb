```python
jumia=requests.get('https://www.jumia.ci/')
soup=BeautifulSoup(jumia.content,'html.parser')
jumia.status_code
```

```python
# Extracting the product name
products=soup.find_all(class_='itm col')
product=[]
for i in products:
    name=i.find(class_='name').get_text()
    lien=i.find('a').get('href')
    lien='https://www.jumia.ci'+lien
    prix=i.find(class_='prc').get_text()
    ancien_prix=i.find(class_='prc').get('data-oprc')
    detail=[name,lien,prix,ancien_prix]
    product.append(detail)
product=pd.DataFrame(product,columns=['name','lien','prix','ancien_prix'])
product
```

```python
product.to_csv('jumia.csv',index=False)
```
